### 2. Build FAISS Indexes

```python
import os
from dotenv import load_dotenv
load_dotenv()

from src.config import settings
from src.ingest import load_documents, make_chunks
from src.rag import build_faiss_store, save_store

docs = load_documents(settings.data_dir)
print("Loaded pages/files:", len(docs))

chunks = make_chunks(docs, chunk_size=settings.chunk_size, overlap=settings.chunk_overlap)
print("Total chunks:", len(chunks))
print("Example chunk:", chunks[0]["source"], chunks[0]["page"], chunks[0]["text"][:120])

store, embedder = build_faiss_store(
    chunks=chunks,
    embed_model_name=settings.embed_model,
    normalize=True,
)

save_store(store, settings.artifacts_dir)
print("Saved artifacts to:", settings.artifacts_dir)
```
